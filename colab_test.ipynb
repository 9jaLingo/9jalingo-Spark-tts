```python
# ============================================================
# Cell 1: GPU check + clone repos + install dependencies
# ============================================================
!nvidia-smi

import os
os.chdir('/content')

# Clone 9jaLingo TTS repo (contains src/inference.py)
if not os.path.exists('/content/9jalingo-Spark-tts'):
    !git clone https://github.com/9jalingo/9jalingo-Spark-tts.git
else:
    print('9jalingo-Spark-tts already cloned')

# Clone SparkVox (BiCodec tokenizer code, required by inference.py)
if not os.path.exists('/content/SparkVox'):
    !git clone https://github.com/SparkAudio/SparkVox.git
else:
    print('SparkVox already cloned')

# Install all required packages
%pip install -q lightning pytorch-lightning hydra-core omegaconf
%pip install -q librosa soundfile torchaudio einops
%pip install -q transformers huggingface_hub
%pip install -q vector-quantize-pytorch descript-audiotools
%pip install -q protobuf==3.20.3 safetensors accelerate sentencepiece

import torch
print(f'\nPyTorch: {torch.__version__}')
print(f'CUDA available: {torch.cuda.is_available()}')
if torch.cuda.is_available():
    print(f'GPU: {torch.cuda.get_device_name(0)}')
print('\nDone.')
```

```python
# ============================================================
# Cell 2: Download model from HuggingFace
# ============================================================
from huggingface_hub import snapshot_download

MODEL_DIR = '/content/Spark-TTS-African'

if not os.path.exists(MODEL_DIR):
    print('Downloading model weights from HuggingFace...')
    snapshot_download(
        repo_id='9jaLingo/9jaLingo-TTS-African-ckpt-25k',
        local_dir=MODEL_DIR,
        local_dir_use_symlinks=False,
    )
else:
    print('Model already downloaded')

# Quick sanity check
for item in sorted(os.listdir(MODEL_DIR)):
    full = os.path.join(MODEL_DIR, item)
    if os.path.isdir(full):
        print(f'  {item}/ ({len(os.listdir(full))} files)')
    else:
        size = os.path.getsize(full)
        label = f'{size/1e6:.1f} MB' if size > 1e6 else f'{size/1024:.0f} KB'
        print(f'  {item} ({label})')
print('\nModel ready.')
```

```python
# ============================================================
# Cell 3: Load the TTS engine from src/inference.py
# ============================================================
import sys

# SparkVox must be on the path (BiCodec lives there)
SPARKVOX_DIR = '/content/SparkVox'
REPO_DIR = '/content/9jalingo-Spark-tts'

if SPARKVOX_DIR not in sys.path:
    sys.path.insert(0, SPARKVOX_DIR)
if REPO_DIR not in sys.path:
    sys.path.insert(0, REPO_DIR)

os.chdir(SPARKVOX_DIR)

from src.inference import NaijaLingoTTS

tts = NaijaLingoTTS(
    model_dir=MODEL_DIR,
    sparkvox_dir=SPARKVOX_DIR,
)
print('\nTTS engine ready!')
```

```python
# ============================================================
# Cell 4: TEST 1 — Standard TTS (cached speaker)
# ============================================================
# No audio upload needed. Pick a speaker_id from the 325-speaker cache.
# Change SPEAKER_ID and TEXT to test different languages.
# ============================================================
from IPython.display import Audio, display

# ======= EDIT THESE =======
SPEAKER_ID = 'hau_f_915'     # Hausa female — change to any speaker from list below
TEXT = 'Rayuwa tafiya ce mai cike da darussa, farin ciki da kalubale. Mutum yana koyo daga kuskure yana girma da hakuri.'
# ==========================

# Browse speakers by language/gender:
# tts.list_speakers(language='hau')   # Hausa
# tts.list_speakers(language='ig')    # Igbo
# tts.list_speakers(language='yor')   # Yoruba
# tts.list_speakers(language='pcm')   # Pidgin

info = tts._speaker_cache.get(SPEAKER_ID, {})
print(f'Speaker : {SPEAKER_ID} | {info.get("language","?")} | {info.get("gender","?")} | {info.get("num_samples","?")} samples')
print(f'Text    : {TEXT}')

wav_standard = tts.speak(TEXT, speaker_id=SPEAKER_ID, temperature=0.8)
tts.save(wav_standard, '/content/test_standard.wav')

print('\nListening:')
display(Audio('/content/test_standard.wav', autoplay=True))
```

```python
# ============================================================
# Cell 5: TEST 2 — Voice Cloning
# ============================================================
# Uses the audio from Cell 4 as the reference voice,
# then generates new text in that same voice.
#
# To clone your own voice: upload a .wav file and set
# REFERENCE_AUDIO to its path instead.
# ============================================================

# ======= EDIT THESE =======
REFERENCE_AUDIO = '/content/test_standard.wav'   # use output from Cell 4
CLONE_TEXT = 'Mun yi farin ciki sosai da ganin ku a nan yau. Allah ya saka da alheri.'
# ==========================

# To upload your own reference instead:
# from google.colab import files
# uploaded = files.upload()
# REFERENCE_AUDIO = '/content/' + list(uploaded.keys())[0]

print(f'Reference: {REFERENCE_AUDIO}')
print(f'Text     : {CLONE_TEXT}')

wav_cloned = tts.clone(CLONE_TEXT, reference_audio=REFERENCE_AUDIO, temperature=0.8)
tts.save(wav_cloned, '/content/test_cloned.wav')

print('\nOriginal reference voice:')
display(Audio(REFERENCE_AUDIO, autoplay=False))

print('Cloned output:')
display(Audio('/content/test_cloned.wav', autoplay=True))
```
